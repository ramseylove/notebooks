```python
def logger(msg):
    def log_message():
        print('Log:', msg)
        
    return log_message
```

```python
log_hi = logger('fart')
log_hi()
```

```python
def html_tag(tag):
    def wrap_text(msg):
        print('<{0}>{1}</{0}>'.format(tag, msg))
        
    return wrap_text
```

```python
print_h1 = html_tag('h1')
print_h1('Test Headline!')
print_h1('Another Headline!')
```

```python
#FizzBuzz
for num in range(1101):
    if num % 5 == 0 and num % 3 ==0:
        print("FizzBuzz")
    elif num % 3 == 0:
        print("Fizz")
    elif num % 5 == 0:
        print("Buzz")
    else:
        print(num)
```

```python
#Comprehensions create list
def createList(r1, r2):
    return list(range(r1, r2+1))

nums = createList(1, 30)

nums
```

```python
my_list =  [num for num in range(1,30)]
print(my_list)
```

```python
d_list = [num*2 for num in nums]
d_list
```

```python
e_list = [n for n in nums if n%2 ==0]
e_list
```

```python
#lamdas # map returns map object which is an iterator
m_list = map(lambda n: n*n, nums)
print(list(m_list))
```

```python
# filter returns a fliter object
f_list = filter(lambda n: n%2 == 0, nums)
list(f_list)
```
