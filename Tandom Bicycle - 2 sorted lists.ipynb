```python
def tandemBicycle(redShirtSpeeds, blueShirtSpeeds, fastest):
    redShirtSpeeds.sort(reverse=fastest)
    blueShirtSpeeds.sort()
    
    total = 0

    for i in range(len(redShirtSpeeds)):
        pair = [redShirtSpeeds[i], blueShirtSpeeds[i]]
        pair.sort()
        total += pair[1]
    return total
```
